```python
import json
from datetime import datetime

# file paths
input_file = "data/nz_trip/trip.json"
points_out = "data/nz_trip/points.geojson"
line_out = "data/nz_trip/route.geojson"

# load
with open(input_file, "r", encoding="utf8") as f:
    data = json.load(f)

steps = data["all_steps"]

features = []
coords = []

for step in steps:
    loc = step.get("location", {})
    lat = loc.get("lat")
    lon = loc.get("lon")
    name = step.get("display_name", "")
    start_time = step.get("start_time")

    if lat is None or lon is None:
        continue

    coords.append([lon, lat])

    features.append({
        "type": "Feature",
        "geometry": {"type": "Point", "coordinates": [lon, lat]},
        "properties": {
            "name": name,
            "timestamp": start_time
        }
    })

# write points
with open(points_out, "w", encoding="utf8") as f:
    json.dump({"type": "FeatureCollection", "features": features}, f, indent=2)

# write route line
line = {
    "type": "Feature",
    "geometry": {"type": "LineString", "coordinates": coords},
    "properties": {"name": "Trip Route"}
}

with open(line_out, "w", encoding="utf8") as f:
    json.dump({
        "type": "FeatureCollection",
        "features": [line]
    }, f, indent=2)

print("Done!")
```
